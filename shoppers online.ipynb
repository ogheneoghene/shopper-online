```python
categorical_vars = ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend']
for var in categorical_vars:
    sns.countplot(df[var])
    plt.title(f'Distribution of {var}')
    plt.xticks(rotation=90)
    plt.show()
```

* This to show that 2 have the highest count of operating system ,while 5 have the lowest operating system.
* Two have the highest distribution of browser,while nothing for 9,11 and 12 no browser to show on the chat.
* Region one have the highest number of customer.

```python
conversion_rate = df["Revenue"].value_counts(normalize=True) * 100
print(conversion_rate)
```

```python
monthly_visitors = df["Month"].value_counts().sort_index()
monthly_revenue = df.groupby("Month")["Revenue"].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_visitors.index, y=monthly_visitors.values)
plt.xlabel("Month")
plt.ylabel("Number of Visitors")
plt.title("Number of Visitors by Month")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_revenue.index, y=monthly_revenue.values)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Revenue by Month")
plt.show()
```

```python
plt.scatter(df["BounceRates"], df["ExitRates"])
plt.xlabel("Bounce Rates")
plt.ylabel("Exit Rates")
plt.title("Bounce Rates vs Exit Rates")
plt.show()
```

```python
special_day_revenue = df.groupby("SpecialDay")["Revenue"].mean()

plt.plot(special_day_revenue.index, special_day_revenue.values, marker="o")
plt.xlabel("Special Day")
plt.ylabel("Revenue")
plt.title("Average Revenue by Special Day")
plt.xticks(np.arange(0, 1.1, 0.1))
plt.show()
```
